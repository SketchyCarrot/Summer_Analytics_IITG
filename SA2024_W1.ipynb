### 4. Name the pokemon(s) with Maximum Defense.

```python
max_defense = np.max(df["Defense"])
defense = df["Defense"]
names = df.index

for i in range(len(names)-1):
        if defense[i] == max_defense:
                print(names[i])
```

### 5. Which poison pokemon has the strongest attack ?

```python
attack = df["Attack"]
poison_attacks = []
type1 = df["Type 1"]
type2 = df["Type 2"]


for i in range(len(names)-1):
        if type1[i] == "Poison" or type2[i] == "Poison":
                poison_attacks.append(attack[i])

max_poison_attack = np.max(poison_attacks)
for i in range(0, len(names)-1):
        if type1[i] == "Poison" or type2[i] == "Poison":
                if attack[i] == max_poison_attack:
                        print(names[i])
```
